```python
import polars as pl
import pandas as pd
import os
```

```python
import gc 
gc.collect()
```

```python
train_path = "/kaggle/input/leap-atmospheric-physics-ai-climsim/train.csv"
sample_ratio = 0.16
batch_size = 50000

reader = pl.read_csv_batched(
    train_path,
    batch_size=batch_size,
)
```

```python
batches = reader.next_batches(5)
sample_data = []
counter = 0

while batches:
    batches = pl.concat(batches)
    batches = batches.sample(fraction=sample_ratio, shuffle=True)
    sample_data.append(batches)
    
    counter += len(batches)
    print(counter)
    
    batches = reader.next_batches(5)
```

```python
sample_data = pl.concat(sample_data)
sample_data = sample_data.sample(1500000, shuffle=True)
sample_data.write_parquet("train_sampled_1_n_half_m.parquet")
```
